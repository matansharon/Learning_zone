```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# get titanic & test csv files as a DataFrame
titanic_df = pd.read_csv("titanic_train.csv")
test_df= pd.read_csv("titanic_test.csv")
# preview the data
titanic_df.head()
test_df.head()
titanic_df.info()
print("----------------------------")
test_df.info()
titanic_df = titanic_df.drop(['PassengerId','Name','Ticket'], axis=1)
test_df    = test_df.drop(['Name','Ticket'], axis=1)
titanic_df.head()   
test_df.head()
titanic_df["Embarked"] = titanic_df["Embarked"].fillna("S")
test_df["Embarked"] = test_df["Embarked"].fillna("S")
titanic_df.info()
print("----------------------------")
test_df.info()
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df["Age"].median())
test_df["Age"] = test_df["Age"].fillna(test_df["Age"].median())
titanic_df.info()
print("----------------------------")
test_df.info()
titanic_df["Fare"] = titanic_df["Fare"].fillna(titanic_df["Fare"].median())
test_df["Fare"] = test_df["Fare"].fillna(test_df["Fare"].median())
titanic_df.info()
print("----------------------------")
test_df.info()
titanic_df["FamilySize"] = titanic_df["SibSp"] + titanic_df["Parch"] + 1
test_df["FamilySize"] = test_df["SibSp"] + test_df["Parch"] + 1
titanic_df["IsAlone"] = 0
test_df["IsAlone"] = 0
titanic_df.loc[titanic_df["FamilySize"] == 1, "IsAlone"] = 1
test_df.loc[test_df["FamilySize"] == 1, "IsAlone"] = 1
titanic_df = titanic_df.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
test_df    = test_df.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
titanic_df.head()
test_df.head()
titanic_df["Age*Class"] = titanic_df.Age * titanic_df.Pclass
test_df["Age*Class"] = test_df.Age * test_df.Pclass
titanic_df.head()
test_df.head()
titanic_df = titanic_df.drop(['Cabin'], axis=1)
test_df    = test_df.drop(['Cabin'], axis=1)
titanic_df.head()
test_df.head()
titanic_df = titanic_df.drop(['Embarked'], axis=1)
test_df    = test_df.drop(['Embarked'], axis=1)
print(titanic_df.head())
test_df.head()
#convert the sex column in  to numeric values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
X = titanic_df.drop("Survived", axis=1)
Y = titanic_df["Survived"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
coeff_df = pd.DataFrame(titanic_df.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])
coeff_df.sort_values(by='Correlation', ascending=False)

```
