# Scatter Plots and Trend Lines

Scatter plots, also known as scatter graphs or charts, are used to show the relationship between two dimensions or measurements of a group. The independent variable is typically placed along the horizontal axis, and the dependent variable is along the vertical axis. While there is often a correlation between the two variables, it does not imply causation.

The trend line, or line of best fit, represents the general direction of the data. It is calculated using the least squares method. The equation for the trend line in a scatter plot follows the form \( y = mx + b \), where \( m \) is the slope, indicating the rate of increase or decrease, and \( b \) is the y-intercept, indicating where the line crosses the y-axis. 

The slope (\( m \)) can be calculated by the formula:

\( m = \frac{n(\sum xy) - (\sum x)(\sum y)}{n(\sum x^2) - (\sum x)^2} \)

And the y-intercept (\( b \)) by the formula:

\( b = \frac{\sum y - m(\sum x)}{n} \)

These calculations can be made easier with software or calculators, which use these formulas to determine the equation of the trend line.






```python

import matplotlib.pyplot as plt
import numpy as np

# Example dataset
item_count = np.array([3, 6, 1, 0, 4, 9, 8, 2, 5, 7])
order_total = np.array([2.08, 5.00, 9.65, 0.00, 16.00, 45.24, 95.00, 3.50, 25.00, 55.00])

# Calculate the trend line
m, b = np.polyfit(item_count, order_total, 1)

# Create scatter plot
plt.scatter(item_count, order_total)

# Add trend line
plt.plot(item_count, m*item_count + b, color='red')

# Label axes
plt.xlabel('Item Count')
plt.ylabel('Order Total ($)')
plt.title('Scatter Plot of Item Count vs Order Total')

# Show plot
plt.show()
```
