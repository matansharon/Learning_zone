# Distribution Plots: Histograms

Histograms are a type of distribution plot that represent the distribution of a continuous variable. They are useful for visualizing the frequency distribution of large datasets.

## Characteristics of Histograms:

- **Horizontal Axis:** Represents a continuous characteristic (e.g., age).
- **Vertical Axis:** Shows the frequency or count (e.g., population in thousands).
- **Bins:** The horizontal axis is divided into intervals known as bins (e.g., age groups 0-9, 10-19, etc.).
- **Continuous Data:** Unlike discrete variables, continuous variables can take on any value within a range, allowing for finer granularity in the data.
- **Bar Gaps:** Histogram bars touch each other, indicating the continuity of the variable.
- **Bin Width:** Each bin should have the same width to ensure consistency in representation.
- **Interpreting Bins:** The height of each bin corresponds to the frequency of the data within that bin.

## Creating Histograms:

To create a histogram, determine the smallest and largest values in the dataset and decide on a reasonable bin width. Then, count the number of occurrences within each bin and plot the bars accordingly.




```python
## Example of a Histogram Plot in Python:

import matplotlib.pyplot as plt

# Example age data in a city
ages = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 2, 102, 95, 85, 22, 35, 30, 9, 45, 55]
#add to ages 100 more random ages with probability of gaussian distribution
import numpy as np
ages = ages + list(np.random.normal(50, 20, 100))

# Creating the histogram
plt.hist(ages, bins=11, edgecolor='black')

# Adding titles and labels
plt.title('Population Distribution by Age')
plt.xlabel('Ages')
plt.ylabel('Population in Thousands')

# Displaying the histogram
plt.show()
```
