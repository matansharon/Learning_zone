# Lecture Summary: Discrete vs. Continuous Random Variables

This lecture covers the fundamental concepts of **discrete and continuous random variables** and their implications for probability calculations.

## Continuous Random Variables

- **Definition**: Variables that represent data which can take on any value within a given range. Examples include dimensions (length, width, height) or the height of people.
- **Characteristics**: Can theoretically assume an infinite number of values between any two points.
- **Example**: The height of people is considered continuous as it can vary infinitely within a range, limited only by measurement precision.

## Discrete Random Variables

- **Definition**: Variables that represent countable outcomes. Examples include the number of children in a family or the number of heads in coin flips.
- **Characteristics**: Can only take specific, distinct values with no in-between values possible.
- **Example**: A family cannot have 2.7 children, indicating the discrete nature of such a variable.

## Probability Calculations

### Discrete Random Variables Example

- **Scenario**: Flipping a coin once, with outcomes being heads or tails, each having a probability of 1/2.
- **Probability Distribution**: Demonstrated through a table and graph, representing the number of heads obtained in a single coin flip with their associated probabilities.

This lecture lays the groundwork for understanding and calculating probabilities for discrete and continuous variables, highlighting the distinct nature of countable outcomes versus variables that can assume an infinite range of values.


```python
import matplotlib.pyplot as plt
import numpy as np

# Discrete Random Variable Example: Coin Flip
# Let's say X represents the number of heads in a single coin flip
outcomes = ['Heads', 'Tails']
probabilities = [0.5, 0.5]

plt.figure(figsize=(10, 5))

# Plot for Discrete Random Variable (Coin Flip)
plt.subplot(1, 2, 1)
plt.bar(outcomes, probabilities, color=['blue', 'orange'])
plt.title('Probability Distribution of Coin Flip')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.ylim(0, 1)

# Continuous Random Variable Example: Hypothetical Variable
# Assuming a normal distribution for demonstration purposes
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Plot for Continuous Random Variable
plt.subplot(1, 2, 2)
count, bins, ignored = plt.hist(s, 30, density=True, alpha=0.6, color='g')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.title('Probability Distribution of a Continuous Variable')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```
